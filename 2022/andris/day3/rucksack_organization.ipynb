### Day 3, Part 1

```python
RUCKSACK_FILE = "day3_input1.txt"

def get_priority(item: str) -> int:
    if item.islower():
        return ord(item) - ord("a") + 1
    else:
        return ord(item) - ord("A") + 27

with open(RUCKSACK_FILE, "r") as f:
    rucksacks = [line.strip() for line in f.readlines()]

duplicate_priority_sum = 0
for rucksack in rucksacks:
    half_idx = int(len(rucksack) / 2)
    compartment_1 = rucksack[:half_idx]
    compartment_2 = rucksack[half_idx:]

    for item in compartment_1:
        if item in compartment_2:
            duplicate_item = item
            break

    duplicate_priority_sum += get_priority(duplicate_item)
    print(f"{rucksack} : {compartment_1} | {compartment_2} : {duplicate_item} ({get_priority(duplicate_item)})")

print(f"Duplicate priority sum: {duplicate_priority_sum}")
```

### Day 3, Part 2

```python
badge_priority_sum = 0
for idx in range(0, len(rucksacks), 3):
    badge_item = None
    for item in rucksacks[idx]:
        if item in rucksacks[idx + 1] and item in rucksacks[idx + 2]:
            badge_item = item
            break
    if not badge_item:
        raise ValueError("Didn't find badge!")
        
    badge_priority_sum += get_priority(badge_item)
    print(f"Group: [{','.join(rucksacks[idx:idx+3])}] | Badge: {badge_item} ({get_priority(badge_item)}")

print(f"Badge priority sum: {badge_priority_sum}")
```
