### Day 5, Part 1

```python
SUPPLY_STACK_INSTRUCTIONS_FILE = "day5_input1_instructions.txt"

SUPPLY_STACKS_START = [
    "DLJRVDF",
    "TPMBVHJS",
    "VHMFDGPC",
    "MDPNGQ",
    "JLHNF",
    "NFVQDGTZ",
    "FDBL",
    "MJBSVDN",
    "GLD"
]

with open(SUPPLY_STACK_INSTRUCTIONS_FILE, "r") as f:
    instructions = f.readlines()

supply_stacks = SUPPLY_STACKS_START.copy()

print(f"Starting supply stacks: {supply_stacks}")
for instruction in instructions:
    instruction_parts = instruction.split()
    num_to_move = int(instruction_parts[1])
    source = int(instruction_parts[3])
    destination = int(instruction_parts[5])

    supply_stacks[destination - 1] += supply_stacks[source - 1][-num_to_move:][::-1] # move one-by-one
    supply_stacks[source - 1] = supply_stacks[source - 1][:-num_to_move]
    print(f"Move {num_to_move} from {source} to {destination}: {supply_stacks}")

print(f"Final supply stacks: {supply_stacks}")

top_of_stacks = ''.join([stack[-1] for stack in supply_stacks])
print(f"Top of stacks: {top_of_stacks}")
```

```python
supply_stacks = SUPPLY_STACKS_START.copy()

print(f"Starting supply stacks: {supply_stacks}")
for instruction in instructions:
    instruction_parts = instruction.split()
    num_to_move = int(instruction_parts[1])
    source = int(instruction_parts[3])
    destination = int(instruction_parts[5])

    supply_stacks[destination - 1] += supply_stacks[source - 1][-num_to_move:] # move as a group
    supply_stacks[source - 1] = supply_stacks[source - 1][:-num_to_move]
    print(f"Move {num_to_move} from {source} to {destination}: {supply_stacks}")

print(f"Final supply stacks: {supply_stacks}")

top_of_stacks = ''.join([stack[-1] for stack in supply_stacks])
print(f"Top of stacks: {top_of_stacks}")
```
